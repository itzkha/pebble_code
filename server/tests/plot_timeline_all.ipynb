```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import normalizing_tools as tools
import os
import datetime
import pandas as pd
from matplotlib import gridspec
import matplotlib as mpl

users = os.listdir('/media/hector/DATA/articleMoodZackData/')
#users = os.listdir("/media/hector/LaCie/backup/articleMoodZackData/")
min_uploads = 2
activities_unique = ["Commuting", "Eat/Drink", "Education", "Household", "No activity", "Personal care", "Prof. services", "Shopping", "Social/Leisure", "Sports/Active", "Working"]
activity_labels = ["Commuting", "Eat/Drink", "Education", "Household", "Personal care", "Prof. services", "Shopping", "Social/Leisure", "Sports/Active", "Working"]
```

```python
#read all the files
activity_all_users = {}
for this_user in users:
    print 'processing ', this_user
    activities_all = tools.read_user_timeline(this_user)
    if len(activities_all) > min_uploads:
        activity_all_users[this_user] = activities_all
```

```python
#plot timelines for each user
fig = plt.figure(figsize=(15,45))
for i, (this_user, this_timeline) in enumerate(activity_all_users.items()):
    print 'processing ', this_user
    activity_matrix = tools.create_activity_matrix(this_timeline, activities_unique)
    
    ax = fig.add_subplot(len(activity_all_users), 1, i+1)
    ax.matshow(activity_matrix.T, aspect='auto', interpolation='nearest', cmap=cm.binary)
    plt.yticks(range(len(activity_labels)), activity_labels)
    plt.title(this_user)
    days = range(0, activity_matrix.shape[0], 24*60)
    for d in days:
        plt.axvline(d)
    plt.xticks(days, [d/(24*60) for d in days])
    ax.xaxis.set_ticks_position('bottom')
    ax.set_xlabel('Days')
    plt.grid()
plt.tight_layout()
fig.savefig('activity_matrix_all_users.png', dpi=300)
fig.savefig('activity_matrix_all_users.pdf', format='pdf')
```

```python
#plot time spent (logged) in each activity
fig = plt.figure(figsize=(5,45))
for i, (this_user, this_timeline) in enumerate(activity_all_users.items()):
    print 'processing ', this_user
    activity_matrix = tools.create_activity_matrix(this_timeline, activities_unique)
    
    fig.add_subplot(len(activity_all_users), 1, i+1)
    plt.bar(range(activity_matrix.shape[1]), sum(activity_matrix, 0))
    plt.xticks(np.arange(len(activity_labels))+0.5, activity_labels, rotation='vertical')
    plt.ylabel('Time [min]')
    plt.title(this_user)
    plt.grid()
plt.tight_layout()
fig.savefig('activity_bar_all_users.png', dpi=300)
fig.savefig('activity_bar_all_users.pdf', format='pdf')
```

```python
#plot activity histogram per minute

#cmap = plt.get_cmap('terrain')
cmap = plt.get_cmap('Paired')
colors = [cmap(i) for i in np.arange(0,1.1,0.1)]

xticks_pos = range(0, 24 * 60, 60)
xticks = [datetime.time(int(m/60), m%60).strftime('%H:%S') for m in xticks_pos]

fig = plt.figure(figsize=(20,25))
for i, (this_user, this_timeline) in enumerate(activity_all_users.items()):
    print 'processing ', this_user
    activity_matrix = tools.create_activity_matrix(this_timeline, activities_unique)
    
    ax = fig.add_subplot(len(activity_all_users)/2, 2, i+1)
    one_day = np.zeros(((24 * 60), activity_matrix.shape[1]))
    for m in range(activity_matrix.shape[0]):
        one_day[m % (24 * 60),:] += activity_matrix[m,:]
    for m in range(one_day.shape[0]):
        one_day[m,:] /= sum(one_day[m,:])
        one_day[m,:] *= 100

    sp = ax.stackplot(range(one_day.shape[0]), one_day.T, colors=colors, linewidth=0, edgecolor='None')
    ax.set_ylabel('Amount of time [%]')
    ax.margins(0, 0)
    plt.title(this_user)
    plt.xticks(xticks_pos, xticks, rotation='vertical')
proxy = [mpl.patches.Rectangle((0,0), 0,0, facecolor=pol.get_facecolor()[0]) for pol in sp]
lgd_u = fig.legend(proxy, tuple(activity_labels), loc="upper center", bbox_to_anchor=(0.0, 1.005, 1.0, 0.01), ncol=10, mode="expand")
plt.tight_layout()

fig.savefig('activity_hist_all_users_2col.png', dpi=300, bbox_extra_artists=(lgd_u, ), bbox_inches='tight')
fig.savefig('activity_hist_all_users_2col.pdf', format='pdf', bbox_extra_artists=(lgd_u, ), bbox_inches='tight')
```

```python
#plot time spent (logged) in each activity
all_counts = pd.DataFrame()
all_counts_norm = pd.DataFrame()
for i, (this_user, this_timeline) in enumerate(activity_all_users.items()):
    print 'processing ', this_user
    activity_matrix = tools.create_activity_matrix(this_timeline, activities_unique)
    counts = sum(activity_matrix)
    counts_norm = counts / int(activity_matrix.shape[0] / (24*60))
    all_counts = pd.concat([all_counts, pd.DataFrame([counts], columns=activity_labels)], ignore_index=True)
    all_counts_norm = pd.concat([all_counts_norm, pd.DataFrame([counts_norm], columns=activity_labels)], ignore_index=True)

fig = plt.figure(figsize=(6,2))
all_counts.boxplot(rot=90, return_type='dict')
lab = plt.ylabel('Time [min]')
plt.title("Time spent in each activity")
fig.savefig('activity_time_all_users.png', dpi=300, bbox_extra_artists=(lab,), bbox_inches='tight')
fig.savefig('activity_time_all_users.pdf', format='pdf', bbox_extra_artists=(lab,), bbox_inches='tight')

fig = plt.figure(figsize=(6,2))
all_counts_norm.boxplot(rot=90, return_type='dict')
lab = plt.ylabel('Time [min/Day]')
plt.title("Time spent in each activity")
fig.savefig('activity_time_norm_all_users.png', dpi=300, bbox_extra_artists=(lab,), bbox_inches='tight')
fig.savefig('activity_time_norm_all_users.pdf', format='pdf', bbox_extra_artists=(lab,), bbox_inches='tight')
```
