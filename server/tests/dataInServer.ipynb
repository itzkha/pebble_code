```python
%matplotlib inline
import paramiko
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
ssh = paramiko.SSHClient()
# automatically add keys without requiring human intervention
ssh.set_missing_host_key_policy( paramiko.AutoAddPolicy() )
ssh.connect('193.134.218.36', username='smartroot', password='dexterDEXTER0')

stdin, stdout, stderr = ssh.exec_command('cd smartdays/server/uploads; ls')

phones = [p[:-1] for p in stdout.readlines()]
#print phones

all_file_names = []
all_file_dates = []
all_file_sizes = []
base_command = 'cd smartdays/server/uploads'
for dir_name in phones:
    print 'Reading ', dir_name
    stdin, stdout, stderr = ssh.exec_command(base_command + '/'+ dir_name + '; du *')
    size_name = stdout.readlines();
    this_sizes = [f.split('\t')[0] for f in size_name]
    this_names = [f.split('\t')[1][:-1] for f in size_name]
    this_dates = [name.split('_')[2][0:8] for name in this_names]
    all_file_names.append(this_names)
    all_file_sizes.append(this_sizes)
    all_file_dates.append(this_dates)

ssh.close()

dates_unique = list(set([item for sublist in all_file_dates for item in sublist]))
dates_unique.sort()
```

```python
file_counter= [len(f) for f in all_file_names]

x = np.arange(len(file_counter))
plt.bar(x, file_counter)
plt.ylabel('Number of files')
plt.title('User statistics')
_ = plt.xticks(x + 0.6, phones, rotation = 'vertical')
```

```python
file_size = [sum([int(size) for size in file_sizes]) for file_sizes in all_file_sizes]

x = np.arange(len(file_size))
plt.bar(x, file_size)
plt.ylabel('Size of files [Bytes]')
plt.title('User statistics')
_ = plt.xticks(x + 0.6, phones, rotation = 'vertical')
```

```python
files_per_date = np.zeros((len(phones), len(dates_unique)))
for p in range(len(phones)):
    for d in all_file_dates[p]:
        files_per_date[p, dates_unique.index(d)] += 1

fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(files_per_date, interpolation='nearest', cmap=cm.binary)
fig.colorbar(cax)

ticks = np.array(ax.get_xticks(), dtype=int)
ticks = ticks[ticks>=0]
ticks = ticks[ticks<len(dates_unique)]
ax.set_xticks(ticks)
ax.set_xticklabels([dates_unique[t] for t in ticks], rotation = 'vertical')

ticks = np.array(ax.get_yticks(), dtype=int)
ticks = ticks[ticks>=0]
ticks = ticks[ticks<len(phones)]
ax.set_yticks(ticks)
ax.set_yticklabels([phones[t] for t in ticks])


fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.plot(sum(files_per_date))

ticks = np.array(ax.get_xticks(), dtype=int)
ticks = ticks[ticks>=0]
ticks = ticks[ticks<len(dates_unique)]
ax.set_xticks(ticks)
ax.set_xticklabels([dates_unique[t] for t in ticks], rotation = 'vertical')
plt.xlabel('Time')
plt.ylabel('Number of files')
plt.title('User statistics')
```

```python
bytes_per_date = np.zeros((len(phones), len(dates_unique)))
for p in range(len(phones)):
    for d in range(len(all_file_dates[p])):
        bytes_per_date[p, dates_unique.index(all_file_dates[p][d])] += int(all_file_sizes[p][d])

fig = plt.figure()
ax = fig.add_subplot(111)
#cax = ax.matshow(bytes_per_date, interpolation='nearest')
cax = ax.imshow(bytes_per_date, interpolation='nearest', cmap=cm.binary)
fig.colorbar(cax)

ticks = np.array(ax.get_xticks(), dtype=int)
ticks = ticks[ticks>=0]
ticks = ticks[ticks<len(dates_unique)]
ax.set_xticks(ticks)
ax.set_xticklabels([dates_unique[t] for t in ticks], rotation = 'vertical')

ticks = np.array(ax.get_yticks(), dtype=int)
ticks = ticks[ticks>=0]
ticks = ticks[ticks<len(phones)]
ax.set_yticks(ticks)
ax.set_yticklabels([phones[t] for t in ticks])


fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.plot(sum(bytes_per_date))

ticks = np.array(ax.get_xticks(), dtype=int)
ticks = ticks[ticks>=0]
ticks = ticks[ticks<len(dates_unique)]
ax.set_xticks(ticks)
ax.set_xticklabels([dates_unique[t] for t in ticks], rotation = 'vertical')
plt.xlabel('Time')
plt.ylabel('Size of files [Bytes]')
plt.title('User statistics')
```
